```python
import pandas as pd
```

```python
#check if we have a null value
data.isnull().sum()
```

```python
#to check if we have duplicated value. as for this one, we only have 1 duplicated data.
data.duplicated().sum()
```

```python
#to drop the duplicated value. since we only have 1 duplicate, it can't affect the whole dataset.
data.drop_duplicates(inplace = True)
```

```python
#to check if we already updated the dataset
data.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.countplot(x = 'Department', data = data)
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Count of Department')
plt.show()
```

```python
sns.countplot(x = "EducationField" , data = data)
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.title ('Education of Each Employee')
plt.show()
```

```python
educfield = data.groupby("Age")['Education'].sum()
```

```python
educfield
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x = "Age", data = data)
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks()
plt.title('Count of Age')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = "Age", y = "Education", data = data)
plt.xlabel('Age')
plt.ylabel('Education')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (10,10))
educfield.plot.barh()
plt.xlabel('Count')
plt.title('Age and Education')
plt.show()
```

```python
data.sort_values(by='Age' , ascending = False, inplace = True)
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x = 'TotalWorkingYears' , data = data)
plt.xlabel('Years')
plt.ylabel('Age')
plt.title('Total Working Years by Age')
plt.show()
```
