```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/content/drive/MyDrive/Data Analyst Portfolio/netflix.csv'
df = pd.read_csv(path)
```

```python
avg_revs_by_revenue = df.groupby('Subscription Type')['Monthly Revenue'].mean()
```

```python
print(avg_revs_by_revenue )
```

```python
user_by_country = df.groupby('Country')['User ID'].count()
```

```python
print(user_by_country)
```

```python
med_ageby_plan = df.groupby('Plan Duration')['Age'].median()
```

```python
print(med_ageby_plan)
```

```python
total_rev = df.groupby('Gender')['Monthly Revenue'].sum()
```

```python
print(total_rev)
```

```python
 average_ageby_device = df.groupby('Device')['Age'].mean()
```

```python
print(average_ageby_device)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'Age', y = 'Monthly Revenue', hue = 'Gender', data = df)
plt.title('Age vs Monthly Revenue (by Gender)')
plt.xlabel('Age')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
new = df.groupby('Country')['Monthly Revenue'].mean().reset_index
print(new)
```

```python
sns.barplot( x="Country", y="Monthly Revenue", data = df)
plt.title('Average Monthly Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Average Monthly Revenue')
plt.xticks(rotation = 90)
plt.show()
```

```python
n_columns = ['Monthly Revenue' , 'Age']
sns.pairplot(df[n_columns], diag_kind = 'kde')
plt.suptitle('Pairplot of Monthly Revenue and Age', y = 1.02)
plt.show()
```

```python
sns.countplot(x = 'Subscription Type', data = df)
plt.title('Subscription Type Distribution')
plt.xlabel('Subscription Type')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.histplot(df['Monthly Revenue'], bins = 5, kde = True)
plt.title('Monthly Revenue Distribution')
plt.xlabel('Monthly Revenue')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x = 'Device', data = df)
plt.title('Device Distribution')
plt.xlabel('Device')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.histplot (df['Age'], bins = 10, kde = True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
