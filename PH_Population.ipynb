````python
#IMPORT NECESSARY PACKAGES
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
#SET DISPLAY OPTION TO MAX
pd.set_option('display.max_columns', 31)
````

````python
#DIRECTORY OF THE FILE
path = '/content/drive/MyDrive/QGIS PORTFOLIO/PH Pop/25 - 44.shp'
````

````python
#READ THE FILE
ph_data = gpd.read_file(path)
````

````python
#CHECK THE CRS OF THE SHAPEFILE
ph_data.crs
````

````python
#CHECK THE FILE
ph_data.head(10)
````

````python
#DESCRIPTION OF DATA
ph_data.describe()
````

````python
#TYPES ON DATA
ph_data.dtypes
````

````python
#CHECK IF THE DATA HAS NULL VALUES
ph_data.isnull().sum()
````

````python
#CHECK IF THE DATA HAS DUPLICATED
ph_data.duplicated().sum()
````


**FOR EXPLORATORY DATA ANALYSIS**

````python
sns.distplot(ph_data['0 - 24'], kde=True)
plt.show()
````

````python
sns.distplot(ph_data['25 - 44'], kde=True)
plt.show()
````

````python
sns.distplot(ph_data['45 - 69'], kde=True)
plt.show()
````

````python
sns.distplot(ph_data['70 - 80'], kde=True)
plt.show()
````

````python
#AS YOU CAN SEE, THE MISSING NULL VALUES HAS ALREADY APPEARD ON THE DATA.
fig = figsize = (10,15)
youth = ph_data.groupby('NAME_1')['0 - 24'].sum().sort_values(ascending=True).plot(kind='barh', figsize=fig)
plt.ylabel=('Provinces')
````

````python
#AS YOU CAN SEE, THE MISSING NULL VALUES HAS ALREADY APPEARD ON THE DATA.
fig = figsize = (10,15)
working_age = ph_data.groupby('NAME_1')['25 - 44'].sum().sort_values(ascending=True).plot(kind='barh', figsize=fig)
plt.ylabel=('Provinces')
````

````python
#AS YOU CAN SEE, THE MISSING NULL VALUES HAS ALREADY APPEARD ON THE DATA.
fig = figsize = (10,15)
middle_age = ph_data.groupby('NAME_1')['45 - 69'].sum().sort_values(ascending=True).plot(kind='barh', figsize=fig)
plt.ylabel=('Provinces')
````

````python
#AS YOU CAN SEE, THE MISSING NULL VALUES HAS ALREADY APPEARD ON THE DATA.
fig = figsize = (10,15)
old_age = ph_data.groupby('NAME_1')['70 - 80'].sum().sort_values(ascending=True).plot(kind='barh', figsize=fig)
plt.ylabel=('Provinces')
````
